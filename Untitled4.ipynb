```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("athiramolps/ice-cream-sales-and-temperature")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
import os

# List files in the downloaded dataset path
files = os.listdir(path)

# Assuming there's a single CSV file, or picking the first one found
csv_file = [f for f in files if f.endswith('.csv')][0]

# Construct the full path to the CSV file
csv_path = os.path.join(path, csv_file)

# Read the CSV into a pandas DataFrame
df = pd.read_csv(csv_path)

print(f"Successfully loaded '{csv_file}' into a DataFrame named 'df'.")
print("First 5 rows of the DataFrame:")
print(df.head())
```

### Linear Regression Analysis

We will perform a simple linear regression to understand the relationship between 'Temperature (Celsius)' and 'Ice Cream Sales'.

**Steps:**
1.  Define the independent variable (X) and the dependent variable (y).
2.  Import the `LinearRegression` model from `sklearn.linear_model`.
3.  Train the model.
4.  Display the model's coefficients and intercept.
5.  Visualize the regression line.

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Define independent variable (X) and dependent variable (y)
X = df[['Temperature (Celsius)']]  # Independent variable (feature)
y = df['Ice Cream Sales']           # Dependent variable (target)

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Display the coefficients and intercept
print(f"Coefficient: {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")
```

```python
# Make predictions
y_pred = model.predict(X)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Sales')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Ice Cream Sales vs. Temperature with Linear Regression')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Ice Cream Sales')
plt.legend()
plt.grid(True)
plt.show()
```
